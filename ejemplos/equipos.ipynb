Vamos a generar los equipos en forma aleatoria para los proyectos del curso de Aprendizaje Automático Aplicado de la Maestría en Ciencia de Datos.

```python
lista_estudiantes = [
    'ALONSO MOZQUEDA JAVIER ALEJANDRO',
    'CASTRO LERMA MARIO ALEJANDRO',	
    'CAZARES MINJARES JOSE DIEGO',
    'FLORES BRIONES JESUS ANTONIO',	
    'GALVAN LARA KEVIN MANUEL',
    'GUERRERO ESPINOZA MIGUEL ANGEL',	
    'HERRERA RENDON MARCEL',
    'MORENO ESPARZA JUAN ANTONIO',	
    'NORIEGA OCHOA LUIS ANGEL',	
    'ORTEGA SOTO FRANCISCO MANUEL',	
    'ORTEGA TANORI ESTHEFANIA',	
    'PEREZ DIAZ JOEL ALFONSO',
    'RAMIREZ CAMACHO BRAYAN ALEXIS',	
    'RAMIREZ RAMIREZ JEHU JONATHAN',	
    'ROBLES ROBLES GABRIEL SANTIAGO',	
    'SENDAY BEJARANO OSCAR DANIEL',	
    'TIRADO LOPEZ ISAUL',
    'VEGA ROBLES SEBASTIAN ISAAC'
]
```

Vamos a seleccionar entre todos una semilla y la vamos a usar para generar los equipos en forma aleatoria.

```python
import random

# Seleccionar una semilla
semilla = 140970  
random.seed(semilla)
print(f"Semilla seleccionada: {semilla}\n")

# Generar una copia de la lista y barajar pseuroaleatoriamente
l_e = lista_estudiantes.copy()
random.shuffle(l_e)

# Crear 6 equipos de 3 integrantes
equipos = [l_e[i * 3:(i + 1) * 3] for i in range(6)]

# Hacer un string formateado en markdown con los equipos
cadena = ' \n'
for i, equipo in enumerate(equipos):
    cadena += f"- **Equipo {i + 1}**: \n"
    for integrante in equipo:
        cadena += f"  - {integrante}\n"
    cadena + ' \n'
```

```python
from IPython.display import display, Markdown
display(Markdown(cadena))
```
